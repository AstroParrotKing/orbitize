```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
#plot of time/niter as a function of orbit fraction coverage for sma fixed tests
times = np.array([4894.78, 1039.94, 27.49, 36.29, 299.99, 12.70, 25.18, 143.85])
fracs = np.array([15, 35, 45, 55, 65, 75, 85, 95])
niter = np.array([22718, 21918, 23060, 22406, 22123, 21915, 21897, 21612])

plt.plot(fracs, times, '.-')
plt.title('Nested Sampler Convergence Time with Varying Orbital Fraction')
plt.ylabel('Time (min)')
plt.xlabel('Orbit Coverage Percentage')
plt.figure()

plt.plot(fracs, niter, '.-')
plt.title('Nested Sampler Iteration Number with Varying Orbital Fraction')
plt.ylabel('Number of Iterations')
plt.xlabel('Orbit Coverage Percentage')
```

```python
#plots of posteriors for the free sma, orbit frac = 55 run
# csv column order: sma,ecc,inc,argp,pan,tau,plx,mtot
# read in data from csv
eccs = np.zeros(26373)
incs = np.zeros(26373)
taus = np.zeros(26373)
aops = np.zeros(26373)
pans = np.zeros(26373)
plxs = np.zeros(26373)
mtots = np.zeros(26373)
smas = np.zeros(26373)
c = 0

with open('trial_freesma_orbfrac85.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        smas[c] = row[0]
        eccs[c] = row[1]
        incs[c] = row[2]
        aops[c] = row[3]
        pans[c] = row[4]
        taus[c] = row[5]
        plxs[c] = row[6]
        mtots[c] = row[7]
        c += 1


# accepted ground truth values
ecc = 0.5
tau = 0.8
inc = np.pi/4
aop = np.pi/4
pan = np.pi/4
sma = 2.187066262313544
plx = 60.
mtot = 1.2

```

```python
#plotting code for set up box above
plt.figure()
plt.hist(eccs, bins=50)
plt.axvline(x=ecc, color='red')
plt.xlabel('ecc'); plt.ylabel('number of orbits')

plt.figure()
plt.hist(incs, bins=50)
plt.axvline(x=inc, color='red')
plt.xlabel('inc'); plt.ylabel('number of orbits')

plt.figure()
plt.hist(aops, bins=50)
plt.axvline(x=aop, color='red')
plt.xlabel('aop'); plt.ylabel('number of orbits')

plt.figure()
plt.hist(pans, bins=50)
plt.axvline(x=pan, color='red')
plt.xlabel('pan'); plt.ylabel('number of orbits')

plt.figure()
plt.hist(taus, bins=50)
plt.axvline(x=tau, color='red')
plt.xlabel('tau'); plt.ylabel('number of orbits')

plt.figure()
plt.hist(mtots, bins=50)
plt.axvline(x=mtot, color='red')
plt.xlabel('mtot'); plt.ylabel('number of orbits')

plt.figure()
plt.hist(plxs, bins=50)
plt.axvline(x=plx, color='red')
plt.xlabel('plx'); plt.ylabel('number of orbits')

plt.figure()
plt.hist(smas, bins=50)
plt.axvline(x=sma, color='red')
plt.xlabel('sma'); plt.ylabel('number of orbits')
```
