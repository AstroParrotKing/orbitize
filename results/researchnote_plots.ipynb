```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
# read in times and niter as a function of orbit frac from csv
# orbit_frac,exec_time,niter
csv_file = open('samplers/smafixed_data_all.csv')
csv_reader = csv.reader(csv_file, delimiter=',')
row_count = 70
fracs = np.zeros(row_count)
times = np.zeros(row_count)
niters = np.zeros(row_count)
c = 0
        
for row in csv_reader:
    fracs[c] = row[0]
    times[c] = row[1]
    niters[c] = row[2]
    c += 1

# get averages of each orbit frac time to plot as triangles
avg35 = np.mean([933.3578288793564,4271.227384765943,3282.291361598174,
                 2791.691862487793,1638.3739593585333,1678.200587495168,
                 1364.4998216350873,2953.986312063535,1531.0958992878595,
                 4038.456437981129])
avg45 = np.mean([40.793222848574324,45.19875450134278,18.92372397184372,
                 25.51754590670268,25.10114507675171,35.25755203564962,
                 28.464388811588286,124.89631615877151,23.50738483270009,
                 45.512431132793424])
```

```python
#plot of time/niter as a function of orbit fraction coverage for sma fixed tests

plt.plot(fracs, times, '.')
plt.yscale('log')
plt.plot(35, avg35, marker='^')
plt.plot(45, avg45, marker='^')
plt.title('Nested Sampler Convergence Time with Varying Orbital Fraction')
plt.ylabel('Time (min)')
plt.xlabel('Orbit Coverage Percentage')
plt.figure()

plt.plot(fracs, niters, '.')
plt.title('Nested Sampler Iteration Number with Varying Orbital Fraction')
plt.ylabel('Number of Iterations')
plt.xlabel('Orbit Coverage Percentage')
```

```python
# plots of posteriors
# csv column order: sma,ecc,inc,argp,pan,tau,plx,mtot

def read_one(file):
    csv_file = open(file)
    csv_reader = csv.reader(csv_file, delimiter=',')
    row_count = 23076
    eccs = np.zeros(row_count)
    incs = np.zeros(row_count)
    taus = np.zeros(row_count)
    aops = np.zeros(row_count)
    pans = np.zeros(row_count)
    plxs = np.zeros(row_count)
    mtots = np.zeros(row_count)
    smas = np.zeros(row_count)
    c = 0
        
    for row in csv_reader:
        smas[c] = row[0]
        eccs[c] = row[1]
        incs[c] = row[2]
        aops[c] = row[3]
        pans[c] = row[4]
        taus[c] = row[5]
        plxs[c] = row[6]
        mtots[c] = row[7]
        c += 1

    return eccs, incs, taus, aops, pans, plxs, mtots, smas

# accepted ground truth values
ecc = 0.5
tau = 0.8
inc = np.pi/4
aop = np.pi/4
pan = np.pi/4
sma = 2.923475321016234
plx = 60.
mtot = 1.2
```

```python
# add all 10 trials together for 35
file = 'samplers/smafixed_orbfrac35_' 
for i in range(0, 10):
    eccs, incs, taus, aops, pans, plxs, mtots, smas = read_one(file + str(i) + '.csv')
    if i == 0:
        ecc_tot35 = eccs
        inc_tot35 = incs
        tau_tot35 = taus
        aop_tot35 = aops
        pan_tot35 = pans
        plx_tot35 = plxs
        mtot_tot35 = mtots
        sma_tot35 = smas
    ecc_tot35 = np.concatenate((ecc_tot35, eccs))
    inc_tot35 = np.concatenate((inc_tot35, incs))
    tau_tot35 = np.concatenate((tau_tot35, taus))
    aop_tot35 = np.concatenate((aop_tot35, aops))
    pan_tot35 = np.concatenate((pan_tot35, pans))
    plx_tot35 = np.concatenate((plx_tot35, plxs))
    mtot_tot35 = np.concatenate((mtot_tot35, mtots))
    sma_tot35 = np.concatenate((sma_tot35, smas))

print(len(ecc_tot35))
```
